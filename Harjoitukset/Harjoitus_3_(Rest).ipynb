#### Tehtävä 8. Muodosta edellisen tehtävän get-pyynnön tuloksesta DataFrame (koko Junan 1 päivän 06.06.2018 datasta), jos yksikin `scheduledTime`-avaimen arvo on ennen aikaleimaa '2018-06-06T06:00:00.000Z'.

```python
response = requests.get("https://rata.digitraffic.fi/api/v1/train-tracking/2018-06-06/1")

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    aika = pd.Timestamp("2018-06-06T06:00:00.000Z")
    
    muokattu_data = df[df['timestamp'] < aika]
    
    print(muokattu_data.head(5))
else:
    print("Virhe: ", response.status_code)
```

#### Tehtävä 9. Tulosta ainoa rivi generoidusta DataFrame-objektista, jos sellainen muodostetaan.

```python
response = requests.get("https://rata.digitraffic.fi/api/v1/train-tracking/2018-06-06/1")

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    aika = pd.Timestamp("2018-06-06T06:00:00.000Z")
    
    muokattu_data = df[df['timestamp'] < aika]
    
    print(muokattu_data.head(1))
else:
    print("Virhe: ", response.status_code)
```
